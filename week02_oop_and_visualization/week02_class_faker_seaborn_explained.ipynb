## 1. 재현 가능한 고객 샘플 만들기

`random.seed()`를 고정하면 실행할 때마다 같은 결과가 나옵니다.
과제 검증에서는 이렇게 재현성이 있는 코드가 더 안전합니다.


```python
import random
from dataclasses import dataclass

import pandas as pd

random.seed(2026)

names = ["Sungmin", "Minho", "Seoyeon", "Hana", "Jisoo", "Doyun", "Yuna", "Hyun"]
grades = ["basic", "silver", "gold"]


@dataclass
class Customer:
    name: str
    age: int
    grade: str
    spend: int

    def label(self):
        return f"{self.name}({self.grade})"


customers = [
    Customer(
        name=random.choice(names),
        age=random.randint(18, 45),
        grade=random.choice(grades),
        spend=random.randint(30_000, 180_000),
    )
    for _ in range(12)
]

df = pd.DataFrame([customer.__dict__ for customer in customers])
df.head()
```

## 2. 그룹별 통계 확인

데이터 분석에서는 원본 행을 보는 것보다 그룹별 요약을 먼저 보면 패턴을 더 빨리 파악할 수 있습니다.


```python
grade_summary = (
    df.groupby("grade", as_index=False)
    .agg(count=("name", "count"), average_spend=("spend", "mean"), average_age=("age", "mean"))
    .round(1)
    .sort_values("average_spend", ascending=False)
)

grade_summary
```

## 3. 그래프 객체 만들기

그래프는 결과를 눈으로 확인하기 위한 도구입니다.
여기서는 파일 저장 없이 Figure 객체만 만든 뒤 닫아서 검증 환경에서도 안정적으로 실행되게 합니다.


```python
import matplotlib.pyplot as plt
import seaborn as sns

fig, ax = plt.subplots(figsize=(6, 3))
sns.barplot(data=grade_summary, x="grade", y="average_spend", ax=ax)
ax.set_title("Average Spend by Grade")
ax.set_xlabel("Grade")
ax.set_ylabel("Average Spend")
plt.close(fig)

print("등급별 평균 지출 그래프를 생성했습니다.")
```
