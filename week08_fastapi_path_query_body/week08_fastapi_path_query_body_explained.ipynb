## Path, Query, Body 구분
Path는 자원 식별, Query는 조회 조건, Body는 생성/수정할 복합 데이터를 전달할 때 사용합니다.

```python
import pandas as pd

ITEMS = [
    {"id": 1, "name": "rocket", "category": "etl", "price": 10},
    {"id": 2, "name": "webtoon", "category": "crawl", "price": 20},
    {"id": 3, "name": "fastapi", "category": "backend", "price": 30},
]

def read_item_path(item_id: int):
    return next((item for item in ITEMS if item["id"] == item_id), None)

def read_items_query(category=None, limit=10):
    rows = ITEMS
    if category:
        rows = [row for row in rows if row["category"] == category]
    return rows[:limit]

def create_item_body(payload):
    required = {"name", "category", "price"}
    if not required.issubset(payload):
        raise ValueError("missing body fields")
    new_item = {"id": max(item["id"] for item in ITEMS) + 1, **payload}
    ITEMS.append(new_item)
    return new_item

assert read_item_path(1)["name"] == "rocket"
assert len(read_items_query(category="backend")) == 1
assert create_item_body({"name": "jinja", "category": "template", "price": 40})["id"] == 4
pd.DataFrame(ITEMS)
```
