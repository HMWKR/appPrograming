```python
# ============================================================
# 과제 1: faker를 사용하여 1000명의 데이터를 Class로 생성
# ============================================================

from faker import Faker

# --- 1단계: User 클래스 정의 ---
# 강의에서 배운 __init__ 생성자 패턴을 확장하여 8개 속성을 가진 클래스를 만든다.
# 강의의 User('001', 'Tom') 패턴에서 속성을 더 풍부하게 확장한 버전이다.

class User:
    """사용자 정보를 담는 클래스"""
    
    def __init__(self, user_id, name, age, email, address, phone, job, company):
        # self는 생성되는 인스턴스 자기 자신을 가리킨다
        # 각 매개변수를 인스턴스 속성(attribute)으로 바인딩한다
        self.id = user_id       # 사용자 고유 ID (예: '0001')
        self.name = name         # 이름 (예: '김민수')
        self.age = age           # 나이 (예: 25)
        self.email = email       # 이메일 (예: 'minsu@example.com')
        self.address = address   # 주소 (예: '서울특별시 강남구...')
        self.phone = phone       # 전화번호 (예: '010-1234-5678')
        self.job = job           # 직업 (예: '데이터 분석가')
        self.company = company   # 회사명 (예: '삼성전자')


# --- 2단계: Faker로 1000명의 User 객체 생성 ---
# Faker('ko_KR')로 한국어 로케일을 설정하면 한국식 이름, 주소가 생성된다.

fake = Faker('ko_KR')  # 한국어 가짜 데이터 생성기 초기화

users = []  # 1000명의 User 객체를 저장할 리스트

for i in range(1000):
    # 매 반복마다 새로운 User 인스턴스를 생성한다
    # 강의에서 user_1 = User('001', 'Tom')처럼 만들었던 것을
    # faker가 자동으로 값을 채워주는 방식으로 확장한다
    user = User(
        user_id=str(i).zfill(4),              # '0000' ~ '0999' (4자리 문자열)
        name=fake.name(),                      # 한국식 이름 자동 생성
        age=fake.random_int(min=18, max=80),   # 18~80세 랜덤 나이
        email=fake.email(),                    # 이메일 자동 생성
        address=fake.address(),                # 한국식 주소 자동 생성
        phone=fake.phone_number(),             # 전화번호 자동 생성
        job=fake.job(),                        # 직업 자동 생성
        company=fake.company()                 # 회사명 자동 생성
    )
    users.append(user)  # 리스트에 추가

print(f"총 {len(users)}명의 User 객체가 생성되었습니다.")


# --- 3단계: 생성된 데이터 검증 ---
# 처음 5명의 사용자 정보를 출력하여 데이터가 올바르게 생성되었는지 확인한다.

print("=" * 70)
print("처음 5명의 사용자 정보")
print("=" * 70)

for user in users[:5]:
    print(f"\n[ID: {user.id}]")
    print(f"  이름: {user.name}")
    print(f"  나이: {user.age}세")
    print(f"  이메일: {user.email}")
    print(f"  주소: {user.address}")
    print(f"  전화: {user.phone}")
    print(f"  직업: {user.job}")
    print(f"  회사: {user.company}")

print("\n" + "=" * 70)
print(f"\n전체 사용자 ID 목록 (처음 20명): {', '.join([u.id for u in users[:20]])}")
print(f"전체 사용자 이름 목록 (처음 10명): {', '.join([u.name for u in users[:10]])}")
```

```python
# ============================================================
# 과제 2: Tom이 Jerry를 쫓아가는 행위를 method로 구현
# ============================================================

# --- 1단계: User 클래스 정의 (following/followers 기능 포함) ---
# 강의의 Car 클래스에서 accelerate()가 self.speed를 바꾸듯이,
# following_mode()가 self.following과 other.followers를 변경한다.

class User:
    """팔로우 기능을 가진 사용자 클래스"""
    
    def __init__(self, user_id, name):
        self.id = user_id       # 사용자 ID
        self.name = name         # 사용자 이름
        self.following = []      # HAS: 내가 팔로우하는 사람 목록 (속성)
        self.followers = []      # HAS: 나를 팔로우하는 사람 목록 (속성)
    
    # --- 메서드: 팔로우 동작 ---
    # Car 클래스의 accelerate(amount)처럼,
    # 이 메서드는 속성(following, followers)을 변경하는 '행위(Does)'이다.
    def following_mode(self, other):
        """
        self가 other를 팔로우한다.
        - self.following에 other의 이름을 추가
        - other.followers에 self의 이름을 추가
        - 이미 팔로우 중이면 중복 추가하지 않음
        """
        # 중복 팔로우 방지: 이미 팔로우 목록에 있으면 메시지만 출력
        if other.name in self.following:
            print(f"  ⚠ {self.name}은(는) 이미 {other.name}을(를) 팔로우 중입니다.")
            return  # 함수 종료, 아래 코드 실행하지 않음
        
        # 팔로우 관계 설정: self → other
        self.following.append(other.name)    # 내 following 목록에 상대방 추가
        other.followers.append(self.name)    # 상대방 followers 목록에 나를 추가
        
        # 현재 상태 출력
        print(f"  [{len(self.following)}회] {self.name}이(가) {other.name}을(를) 팔로우합니다. "
              f"| {self.name}의 following: {len(self.following)}명 "
              f"| {other.name}의 followers: {len(other.followers)}명")


# --- 2단계: Tom과 Jerry 객체 생성 ---
tom = User('001', 'Tom')
jerry = User('002', 'Jerry')

# --- 3단계: Tom이 Jerry를 10번 following 시도 ---
# 첫 번째는 팔로우가 성공하고, 나머지 9번은 중복으로 무시된다.
print("=" * 70)
print("Tom이 Jerry를 쫓아갑니다! (10번 following 시도)")
print("=" * 70)

for i in range(10):
    print(f"\n[시도 {i+1}/10]")
    tom.following_mode(jerry)
```
