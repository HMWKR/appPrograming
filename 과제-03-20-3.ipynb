### 과제. HTTP server를 사용하여 Client와 Server 코드를 만들어라
- vs code에서 로컬환경에서 py 파일로 2개 코드, sever.py와 clietn.py를 만들어 통신하라

---

## 실행 방법

**1단계: 서버 실행** (터미널 1)
```bash
python server.py
```

**2단계: 클라이언트 실행** (터미널 2)
```bash
python client.py
```

**3단계: 종료** — 서버 터미널에서 `Ctrl+C`

---

| 파일 | 역할 | 사용 모듈 |
|------|------|----------|
| `server.py` | HTTP 서버 (localhost:8080) | `FastAPI` + `uvicorn` (ASGI) |
| `client.py` | HTTP 클라이언트 (GET/POST 요청) | `urllib.request` (표준 라이브러리) |

---

````python
# server.py 내용 확인 (실제 실행은 터미널에서 해야 합니다)
with open('server.py', 'r', encoding='utf-8') as f:
    print(f.read())
````
