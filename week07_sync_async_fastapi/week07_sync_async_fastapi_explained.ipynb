## 동기와 비동기
동기 처리는 하나씩 끝날 때까지 기다리고, 비동기 처리는 기다리는 시간을 다른 작업에 활용합니다.

```python
import asyncio
import time

def sync_job(name, delay=0.05):
    time.sleep(delay)
    return name

async def async_job(name, delay=0.05):
    await asyncio.sleep(delay)
    return name

start = time.perf_counter()
sync_result = [sync_job(i) for i in range(3)]
sync_elapsed = time.perf_counter() - start

start = time.perf_counter()
async_result = await asyncio.gather(*(async_job(i) for i in range(3)))
async_elapsed = time.perf_counter() - start

assert sync_result == async_result
{"sync_elapsed": round(sync_elapsed, 3), "async_elapsed": round(async_elapsed, 3)}
```

## 서버/클라이언트 사고방식
서버 함수는 요청을 받아 응답을 만듭니다. 클라이언트는 그 응답을 호출하고 해석합니다.

```python
PRODUCTS = {1: {"name": "notebook", "price": 3000}, 2: {"name": "pen", "price": 1000}}

def server_get_product(product_id):
    product = PRODUCTS.get(product_id)
    if product is None:
        return {"status": 404, "detail": "not found"}
    return {"status": 200, "data": product}

def client_read(product_id):
    response = server_get_product(product_id)
    return response["data"]["name"] if response["status"] == 200 else "missing"

assert client_read(1) == "notebook"
assert client_read(99) == "missing"
```
